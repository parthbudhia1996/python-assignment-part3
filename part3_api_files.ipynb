```python
import requests
import json
from datetime import datetime
```

```python
# --- TASK 1: FILE READ & WRITE BASICS ---
print("--- Task 1: File Read & Write Basics ---\n")

filename = "python_notes.txt"
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.\n"
    "Topic 2: Lists are ordered and mutable.\n"
    "Topic 3: Dictionaries store key-value pairs.\n"
    "Topic 4: Loops automate repetitive tasks.\n"
    "Topic 5: Exception handling prevents crashes.\n"
]

# Part A - Write:
# 1. Write
with open(filename, "w", encoding="utf-8") as file:
    file.writelines(notes)
# 3. Print
print("File written successfully.")

# 2. Append
with open(filename, "a", encoding="utf-8") as file:
    file.write("Topic 6: File I/O allows persistent data storage.\n")
    file.write("Topic 7: APIs enable communication between systems.\n")
# 3. Print
print("Lines appended.")

# Part B - Read:
# 1. Read
print("\nReading file content:")
with open(filename, "r", encoding="utf-8") as file:
    lines = file.readlines()
    for idx, line in enumerate(lines, 1):
        print(f"{idx}. {line.strip()}")
    # 2. Count
    print(f"\nTotal lines in file: {len(lines)}")

# 3. Keyword Search
search_key = input("\nEnter a keyword to search in notes: ").lower()
found = False
for line in lines:
    if search_key in line.lower():
        print(f"Match found: {line.strip()}")
        found = True
if not found:
    print("No matches found for your keyword.")
```

```python
# --- TASK 2 & 3(PART C): API INTEGRATION & ROBUST API CALLS ---
print("--- Task 2 & 3(PART C): API Integration & Robust API Calls ---\n")

BASE_URL = "https://dummyjson.com/products"

# Step 1: Fetch Products
def fetch_products():
    try:
        response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
        response.raise_for_status() # Trigger error for 4xx/5xx
        data = response.json()

        products = data['products']

        # Step 1: Display Table
        print(f"{'ID':<4} | {'Title':<25} | {'Category':<15} | {'Price':<8} | {'Rating'}")
        print("-" * 70)
        for p in products:
            print(f"{p['id']:<4} | {p['title'][:25]:<25} | {p['category']:<15} | ${p['price']:<7} | {p['rating']}")

        # Step 2: Filter & Sort
        filtered = [p for p in products if p['rating'] >= 4.5]
        filtered.sort(key=lambda x: x['price'], reverse=True)

        print("\nFiltered (Rating >= 4.5) & Sorted by Price:")
        for p in filtered:
            print(f"- {p['title']:<50}: ${p['price']} (Rating: {p['rating']})")

    except requests.exceptions.ConnectionError:
        msg = "Connection failed. Please check your internet."
        print(msg)
    except requests.exceptions.Timeout:
        msg = "Request timed out. Try again later."
        print(msg)
    except Exception as e:
        print(f"Unexpected error: {e}")

fetch_products()

# Step 3: Search by Category
def search_laptops():
    try:
        resp = requests.get(f"{BASE_URL}/category/laptops", timeout=5)
        data = resp.json()
        print("\nLaptops Category:")
        for p in data['products']:
            print(f"Name: {p['title']:<46}, Price: ${p['price']}")
    except requests.exceptions.ConnectionError:
        msg = "Connection failed. Please check your internet."
        print(msg)
    except requests.exceptions.Timeout:
        msg = "Request timed out. Try again later."
        print(msg)
    except Exception as e:
        print(f"Error fetching laptops: {e}")

search_laptops()

# Step 4: POST Request
def add_custom_product():
    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }
    try:
        resp = requests.post(f"{BASE_URL}/add", json=new_product, timeout=5)
        print("\nPOST Response (Simulated):")
        print(resp.json())
    except requests.exceptions.ConnectionError:
        msg = "Connection failed. Please check your internet."
        print(msg)
    except requests.exceptions.Timeout:
        msg = "Request timed out. Try again later."
        print(msg)
    except Exception as e:
        print(f"POST failed: {e}")

add_custom_product()
```

```python
# --- TASK 3: EXCEPTION HANDLING ---
print("--- Task 3: Exception Handling ---\n")

# Part A — Guarded Calculator:
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print(f"Divide 10/2: {safe_divide(10, 2)}")
print(f"Divide 10/0: {safe_divide(10, 0)}")
print(f"Divide 'ten'/2: {safe_divide('ten', 2)}\n")

# Part B — Guarded File Reader:
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        print("File operation attempt complete.")

print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))

# PART C — Robust API Calls:
print("\n--- Part C: Robust API Calls (Update your API code from Task 2)---\n")

# Part D — Input Validation Loop:
while True:
    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ")
    if user_input.lower() == 'quit':
        break

    if not user_input.isdigit() or not (1 <= int(user_input) <= 100):
        print("Invalid input. Please enter a number between 1 and 100.")
        continue

    try:
        resp = requests.get(f"{BASE_URL}/{user_input}")
        if resp.status_code == 404:
            print("Product not found.")
        elif resp.status_code == 200:
            p = resp.json()
            print(f"Result -> Title: {p['title']}, Price: ${p['price']}")
    except Exception as e:
        print(f"Search failed: {e}")
```

```python
# --- TASK 4: LOGGING To FILE ---
print("--- TASK 4: Logging to File ---\n")

def log_error(function_name, error_msg):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_msg}\n"
    with open("error_log.txt", "a") as f:
        f.write(log_entry)

# 1. Trigger ConnectionError
print("Triggering ConnectionError...")
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=2)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError — No connection could be made")

# 2. Trigger HTTP 404
print("Triggering HTTP 404...")
resp = requests.get(f"{BASE_URL}/999")
if resp.status_code != 200:
    log_error("lookup_product", f"HTTPError - {resp.status_code} Not Found for product ID 999")

# Final print of logs:
print("\nContents of error_log.txt:")
try:
    with open("error_log.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Log file empty or not created.")
```
